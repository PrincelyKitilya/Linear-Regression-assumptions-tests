 Checking the validity of Simple Linear regression assumptions using the Palmer Penguins dataset on Seaborn

```python
# Importing the important packages
import pandas as pd
import seaborn as sns
```

```python
# loading dataset 
penguins = sns.load_dataset("penguins")

#checking the first 5 items in the data set 
penguins.head()
```

```python
# checking how the data set looks like how many rows and rows 
penguins.shape
```

There are 7 rows and 344 columns in the dataset 
The columns have variables such as 
    - species
    - island
    - bill_length_mm
    - bill_depth_mm
    - flipper_length_mm
    - body_mass_g
    - sex

There is also some missing data in the dataset as there are NaN Types so we will have to clean them 

```python
#Checking how many types of penguin species there are in the species column
species_type = penguins['species'].unique()
print(species_type)
```

```python
# now i know i have 3 types of penguin species but i don't want to use them all for this analysis
# i want to only use only "Adelie" and "Gentoo" 
#I also want to drop missing values 

penguins_sub = penguins[penguins["species"] != "Chinstrap"] #selecting needed columns
penguins_final = penguins_sub.dropna() #droping NaN from the selected columns
penguins_final.reset_index(inplace=True, drop=True)
```

## Analyzing the Chinstrap Penguins

```python
#importing the correct pacakges
import pandas as pd
import seaborn as sns

#loading dataset 
penguins = sns.load_dataset("Penguins")

#Examining the first 5 rows
penguins.head()
```

```python
# Checking out how many species of Penguins there are 
print(penguins["species"].unique())
```

```python
# isolating the "Chinstrap" penguin species
chinstrap_penguins =penguins[penguins["species"] ==  "Chinstrap"]

# reseting the indexes of the chinstrap penguins
# .reset index method takes in two arguments inplace = boolean(True) and drop = boolean (True)
chinstrap_penguins.reset_index(inplace=True, drop=True)
chinstrap_penguins.head()
```

```python
# looking at the shape of the data 
chinstrap_penguins.shape
```

```python
# Settin up the model for construction 
# isolate the dependent and independent variables 
# we want to access bill_depth_mm and flipper_length_mm
ols_data = chinstrap_penguins[["bill_length_mm", "flipper_length_mm"]]
```

```python
# we are going to use the statsmodels library for our analysis
# we need to write a formula using these variables remember y always on the left side 
# y - dependent variable, x - independent variable
ols_formula = "flipper_length_mm ~ bill_length_mm"
```

```python
# constructing the model 
from statsmodels.formula.api import ols

#Building OLS and fitting model to data 
OLS = ols(formula=ols_formula, data=ols_data)
model = OLS.fit()
model.summary()
```
